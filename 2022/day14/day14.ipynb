```julia
using Pipe
```

```julia
normInf(vec::Tuple{Int64,Int64}) = map(abs, vec) |> Base.splat(max)

normalizeInf(vec::Tuple{Int64,Int64}) = map(x -> sign(x) * min(abs(x), 1), vec)

displayCave(blocked::Set{Tuple{Int64,Int64}}, restedSand::Set{Tuple{Int64,Int64}}, minX::Int64, maxX::Int64, minY::Int64, maxY::Int64) =
  begin
    cave = fill('.', (maxY-minY+1, maxX-minX+1))
    for (x, y) in blocked
      cave[y - minY + 1, x - minX + 1] = '#'
    end
    for (x, y) in restedSand
      cave[y - minY + 1, x - minX + 1] = 'o'
    end

    stringFormVec = Vector{String}()
    for i in 1:(maxY-minY+1)
      strToCreate = ""
      for j in 1:(maxX-minX+1)
        strToCreate *= cave[i, j]
      end
      push!(stringFormVec, strToCreate)
    end

    # display(cave)
    # println(join(stringFormVec, "\n"))
    # display(join(stringFormVec, "\n"))
    display(stringFormVec)
  end
```

```julia
blocked = Set{Tuple{Int64,Int64}}()
maxX = 0
maxY = 0
minX = 1000000000000
minY = 1000000000000
for line in eachline("input.txt")
  splitted = split(line, " -> ")
  xyPairs = @pipe map(x -> split(x, ","), splitted) |> map(x -> Tuple(Int64(y) for y in parse.(Int64, x)), _)
  # xyPairs = @pipe map(x -> split(x, ","), splitted) |> map(x -> (parse(Int64, x[1]), parse(Int64, x[2])), _)
  for i in 2:length(xyPairs)
    step = normalizeInf(xyPairs[i] .- xyPairs[i-1])
    curr = xyPairs[i-1]
    while curr != xyPairs[i]
      push!(blocked, curr)
      curr = curr .+ step
    end
  end

  for (x, y) in xyPairs
    if x > maxX
      maxX = x
    elseif x < minX
      minX = x
    end
    if y > maxY
      maxY = y
    elseif y < minY
      minY = y
    end
  end
end
```

```julia
# task 1
sandStart = (500, 0)
minY = min(minY, sandStart[2])
movement = [(0, 1), (-1, 1), (1, 1)]
sandCount = 0
restedSand = Set{Tuple{Int64,Int64}}()

rested = true
while rested
  p = sandStart
  newP = sandStart
  hasMoved = true

  while Base.between(p[1], minX, maxX) && Base.between(p[2], minY, maxY) && hasMoved
    for fall in movement
      if ((p .+ fall) in blocked || (p .+ fall) in restedSand) |> .!
        newP = p .+ fall
        break
      end
    end
    hasMoved = newP != p
    p = newP
    if !hasMoved
      push!(restedSand, newP)
    end
  end

  if Base.between(sandCount, 610, 640)
    displayCave(blocked, restedSand, minX, maxX, minY, maxY)
  end
  rested = !hasMoved
  sandCount += 1
end

sandCount
```
